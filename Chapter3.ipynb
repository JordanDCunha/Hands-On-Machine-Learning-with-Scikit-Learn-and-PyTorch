# MNIST

In this chapter we will be using the **MNIST dataset**, which is a set of 70,000 small images of digits handwritten by high school students and employees of the US Census Bureau. Each image is labeled with the digit it represents.

This dataset is often called the **“hello world” of machine learning**. Whenever people come up with a new classification algorithm, they are curious to see how it performs on MNIST. As a result, almost everyone who learns machine learning eventually works with this dataset.


```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml("mnist_784", as_frame=False)
```

Scikit-Learn provides several types of dataset-loading functions:

- **fetch_\***: downloads real-world datasets (e.g., `fetch_openml`)
- **load_\***: loads small toy datasets bundled with Scikit-Learn
- **make_\***: generates synthetic datasets for testing

Most datasets are returned as either:
- `(X, y)` NumPy arrays, or
- `Bunch` objects (dictionaries with attribute access)

The MNIST dataset is special because it contains image data. For this reason, we use `as_frame=False` so the data is returned as NumPy arrays instead of Pandas objects.


```python
X, y = mnist.data, mnist.target

X.shape, y.shape
```

There are **70,000 images**, and each image has **784 features**.

This is because each image is **28 × 28 pixels**, and each feature represents a pixel’s intensity:
- `0` → white
- `255` → black


```python
X
```

```python
y
```

Let’s visualize one digit from the dataset. We reshape the feature vector into a 28 × 28 image and display it using Matplotlib.


```python
import matplotlib.pyplot as plt

def plot_digit(image_data):
    image = image_data.reshape(28, 28)
    plt.imshow(image, cmap="binary")
    plt.axis("off")

some_digit = X[0]
plot_digit(some_digit)
plt.show()
```

This looks like a **5**, and indeed the label confirms it.


```python
y[0]
```

To give a sense of the complexity of this task, MNIST contains a wide variety of handwriting styles.

That said, the images are:
- Clean
- Centered
- Not heavily rotated
- Roughly the same size

This means MNIST requires **very little preprocessing** compared to most real-world datasets.


```python
X_train, X_test = X[:60000], X[60000:]
y_train, y_test = y[:60000], y[60000:]

X_train.shape, X_test.shape
```

The dataset is already split into:
- **60,000 training images**
- **10,000 test images**

The training set is also **shuffled**, which is important because:
- Cross-validation folds remain balanced
- Order-sensitive algorithms perform better
- The model does not see many similar digits in a row


# Training a Binary Classifier

To simplify the problem, we will first try to identify just **one digit** — for example, the number **5**.

This “5-detector” is a **binary classifier**, meaning it distinguishes between only two classes:
- **5**
- **not 5**


```python
y_train_5 = (y_train == '5')  # True for all 5s, False for all other digits
y_test_5 = (y_test == '5')
```

Now we need to choose a classifier and train it.

A good starting point is a **Stochastic Gradient Descent (SGD) classifier**. This type of classifier:
- Scales well to large datasets
- Trains instances one at a time
- Is well suited for **online learning**

We will use Scikit-Learn’s `SGDClassifier`.


```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train, y_train_5)
```

The classifier is now trained.

Let’s test it on a single image and see whether it detects a **5**.


```python
sgd_clf.predict([some_digit])
```

The classifier predicts **True**, meaning it believes this image represents a **5**.

In this case, the prediction is correct. However, a single prediction does not tell us much about overall performance. Next, we will evaluate the classifier more rigorously using proper performance metrics.


# Performance Measures

Evaluating a classifier is often trickier than evaluating a regressor, especially when working with **skewed datasets**.

In this section, we will explore:
- Accuracy and why it can be misleading
- Confusion matrices
- Precision, recall, and F1 score
- Precision/recall trade-offs
- ROC curves and AUC


### Measuring Accuracy Using Cross-Validation

Let’s start by evaluating our classifier using **k-fold cross-validation** with 3 folds.


```python
from sklearn.model_selection import cross_val_score

cross_val_score(sgd_clf, X_train, y_train_5, cv=3, scoring="accuracy")
```

The accuracy looks impressive—over 95%!  
But accuracy can be misleading when dealing with **imbalanced datasets**.

Let’s compare this to a dummy classifier that always predicts “not 5”.


```python
from sklearn.dummy import DummyClassifier

dummy_clf = DummyClassifier()
dummy_clf.fit(X_train, y_train_5)
cross_val_score(dummy_clf, X_train, y_train_5, cv=3, scoring="accuracy")
```

Even this useless classifier achieves over **90% accuracy**, simply because most images are not 5s.

This demonstrates why **accuracy is not a reliable metric** for classification when classes are imbalanced.


### Implementing Cross-Validation Manually

Sometimes you need more control than `cross_val_score()` provides.
Below is a manual implementation using `StratifiedKFold`.


```python
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone

skfolds = StratifiedKFold(n_splits=3)

for train_index, test_index in skfolds.split(X_train, y_train_5):
    clone_clf = clone(sgd_clf)

    X_train_folds = X_train[train_index]
    y_train_folds = y_train_5[train_index]
    X_test_fold = X_train[test_index]
    y_test_fold = y_train_5[test_index]

    clone_clf.fit(X_train_folds, y_train_folds)
    y_pred = clone_clf.predict(X_test_fold)

    print(sum(y_pred == y_test_fold) / len(y_pred))
```

### Confusion Matrix

A confusion matrix shows how predictions are distributed across classes.


```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix

y_train_pred = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3)
confusion_matrix(y_train_5, y_train_pred)
```

Each row represents the **actual class**, and each column represents the **predicted class**.

- True Negatives (TN)
- False Positives (FP)
- False Negatives (FN)
- True Positives (TP)


### Precision and Recall

- **Precision**: How many predicted positives are actually positive
- **Recall**: How many actual positives are correctly detected


```python
from sklearn.metrics import precision_score, recall_score

precision_score(y_train_5, y_train_pred)
```

```python
recall_score(y_train_5, y_train_pred)
```

Precision and recall often move in opposite directions.

To balance them, we use the **F1 score**.


```python
from sklearn.metrics import f1_score

f1_score(y_train_5, y_train_pred)
```

### Precision / Recall Trade-Off

The classifier makes decisions based on a **decision threshold**.
Raising the threshold increases precision but decreases recall.


```python
y_scores = sgd_clf.decision_function([some_digit])
y_scores
```

```python
threshold = 3000
(y_scores > threshold)
```

To analyze all thresholds, we compute precision and recall values for every possible decision score.


```python
from sklearn.metrics import precision_recall_curve

y_scores = cross_val_predict(
    sgd_clf, X_train, y_train_5, cv=3, method="decision_function"
)

precisions, recalls, thresholds = precision_recall_curve(y_train_5, y_scores)
```

### ROC Curve

The ROC curve plots:
- True Positive Rate (Recall)
- False Positive Rate


```python
from sklearn.metrics import roc_curve, roc_auc_score

fpr, tpr, thresholds = roc_curve(y_train_5, y_scores)
roc_auc_score(y_train_5, y_scores)
```

### Comparing with a Random Forest Classifier

Let’s compare our SGD classifier with a Random Forest.


```python
from sklearn.ensemble import RandomForestClassifier

forest_clf = RandomForestClassifier(random_state=42)

y_probas_forest = cross_val_predict(
    forest_clf, X_train, y_train_5, cv=3, method="predict_proba"
)

y_scores_forest = y_probas_forest[:, 1]
```

```python
f1_score(y_train_5, y_scores_forest >= 0.5)
```

```python
roc_auc_score(y_train_5, y_scores_forest)
```

The Random Forest classifier:
- Achieves much higher precision and recall
- Has a significantly better ROC AUC score

You now understand how to properly evaluate **binary classifiers** and choose metrics that actually matter.


# Multiclass Classification

Whereas binary classifiers distinguish between two classes, multiclass classifiers
(also called multinomial classifiers) can distinguish between more than two classes.


Some Scikit-Learn classifiers (e.g., LogisticRegression, RandomForestClassifier,
GaussianNB) can handle multiple classes natively.

Others (e.g., SGDClassifier and SVC) are strictly binary classifiers.
However, there are strategies to use multiple binary classifiers to perform
multiclass classification.


### One-versus-the-Rest (OvR)

One approach is to train one binary classifier per class.
Each classifier learns to distinguish one class from all the others.

For digits 0–9, this means training 10 classifiers.
The class whose classifier produces the highest score is selected.


### One-versus-One (OvO)

Another strategy is to train a binary classifier for every pair of classes.

If there are N classes, this requires N × (N − 1) / 2 classifiers.
For MNIST (10 classes), that means 45 classifiers.

OvO can be faster for algorithms that scale poorly with large datasets.


Scikit-Learn automatically selects OvR or OvO depending on the classifier.

Let’s train a Support Vector Machine classifier (SVC).
To keep training time reasonable, we’ll only use the first 2,000 images.


```python
from sklearn.svm import SVC

svm_clf = SVC(random_state=42)
svm_clf.fit(X_train[:2000], y_train[:2000])  # multiclass labels
```

Let’s make a prediction for a single digit.


```python
svm_clf.predict([some_digit])
```

Since there are 10 classes, Scikit-Learn used the OvO strategy internally
and trained 45 binary classifiers.

Calling `decision_function()` returns one score per class.


```python
some_digit_scores = svm_clf.decision_function([some_digit])
some_digit_scores.round(2)
```

The predicted class corresponds to the highest score.


```python
class_id = some_digit_scores.argmax()
class_id
```

Each classifier stores the list of class labels in the `classes_` attribute.


```python
svm_clf.classes_
```

```python
svm_clf.classes_[class_id]
```

### Forcing OvR or OvO Manually

You can explicitly choose a strategy using OneVsRestClassifier or
OneVsOneClassifier.


```python
from sklearn.multiclass import OneVsRestClassifier

ovr_clf = OneVsRestClassifier(SVC(random_state=42))
ovr_clf.fit(X_train[:2000], y_train[:2000])
```

Let’s check the prediction and the number of trained classifiers.


```python
ovr_clf.predict([some_digit])
```

```python
len(ovr_clf.estimators_)
```

# Error Analysis

In a real machine learning project, you would now follow a project checklist:
explore data preparation options, try multiple models, fine-tune hyperparameters,
and automate as much as possible.

Here, we assume that we already have a promising model and want to improve it
by analyzing the types of errors it makes.


### Confusion Matrix

A good starting point is the confusion matrix.

We first generate predictions using cross-validation, then compare the true
labels with the predicted labels.


```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

y_train_pred = cross_val_predict(
    sgd_clf, X_train_scaled, y_train, cv=3
)
```

Let’s visualize the confusion matrix.


```python
ConfusionMatrixDisplay.from_predictions(y_train, y_train_pred)
plt.show()
```

Most predictions lie on the main diagonal, which means they were classified
correctly.

However, raw counts can be misleading. Some digits appear more often than others.
To fix this, we normalize the confusion matrix by row (true labels).


```python
ConfusionMatrixDisplay.from_predictions(
    y_train,
    y_train_pred,
    normalize="true",
    values_format=".0%"
)
plt.show()
```

Now the errors are easier to interpret.

For example, only about 82% of the images of digit 5 were classified correctly.
The most common mistake was misclassifying 5s as 8s.

Note that confusion matrices are not symmetric.


### Focusing on Errors Only

To make misclassifications stand out, we can ignore correct predictions by
assigning them zero weight.


```python
sample_weight = (y_train_pred != y_train)

ConfusionMatrixDisplay.from_predictions(
    y_train,
    y_train_pred,
    sample_weight=sample_weight,
    normalize="true",
    values_format=".0%"
)
plt.show()
```

This visualization highlights where the classifier struggles most.

Many digits are misclassified as 8s, suggesting that the model has difficulty
distinguishing other digits from 8s.

Be careful interpreting percentages here: they represent the distribution of
errors, not the overall misclassification rate.


The confusion matrix can also be normalized by column (predicted labels) using
normalize="pred".

This helps answer questions like:
"Of all images predicted as 8, how many were actually something else?"


### Analyzing Individual Errors

Another useful approach is to examine individual misclassified images.

Let’s compare digits 3 and 5, organizing examples like a confusion matrix.


```python
cl_a, cl_b = '3', '5'

X_aa = X_train[(y_train == cl_a) & (y_train_pred == cl_a)]
X_ab = X_train[(y_train == cl_a) & (y_train_pred == cl_b)]
X_ba = X_train[(y_train == cl_b) & (y_train_pred == cl_a)]
X_bb = X_train[(y_train == cl_b) & (y_train_pred == cl_b)]
```

You can now plot images from X_aa, X_ab, X_ba, and X_bb in a grid resembling
a confusion matrix.

This helps visualize which digits are correctly classified and which are confused.


Many of the misclassified digits are poorly written and even difficult for humans
to recognize.

Others seem obvious to us, but remember that we are using a simple linear model
(SGDClassifier). It assigns weights to pixels and sums them — no spatial awareness,
no rotation invariance.


Digits like 3 and 5 differ only by a few pixels. Small shifts or rotations can
easily confuse the classifier.

One way to reduce these errors is to preprocess the images so they are well
centered and aligned. Another, simpler approach is data augmentation.


### Data Augmentation

By adding slightly shifted or rotated versions of the training images, the model
is forced to learn features that are more robust to such variations.

This technique is called *data augmentation* and will be covered in later chapters.


# Multilabel Classification

So far, each instance has been assigned to exactly one class. However, in some
applications, a classifier may need to output multiple labels for a single
instance.

For example, in face recognition, an image may contain multiple people, so the
classifier should assign one label per person. This type of system is called a
*multilabel classification system*.


To illustrate this idea, let’s build a simple multilabel classifier using the
MNIST dataset.

Each digit will receive two labels:
- Whether the digit is **large** (7, 8, or 9)
- Whether the digit is **odd**


```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

We now create two target labels per image and combine them into a single
multilabel target array.


```python
y_train_large = (y_train >= '7')
y_train_odd = (y_train.astype('int8') % 2 == 1)

y_multilabel = np.c_[y_train_large, y_train_odd]
```

Next, we train a classifier that supports multilabel classification.
The KNeighborsClassifier can handle multiple output labels.


```python
knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train, y_multilabel)
```

Let’s make a prediction for a single digit and inspect the result.


```python
knn_clf.predict([some_digit])
```

The output contains two values:
- The first indicates whether the digit is large
- The second indicates whether the digit is odd

For example, the digit 5 is not large and is odd.


### Evaluating a Multilabel Classifier

There are several ways to evaluate multilabel classifiers. One common approach
is to compute a metric such as the F1 score for each label and then average them.


```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import f1_score

y_train_knn_pred = cross_val_predict(
    knn_clf, X_train, y_multilabel, cv=3
)

f1_score(y_multilabel, y_train_knn_pred, average="macro")
```

Using `average="macro"` gives equal weight to each label.

If some labels are more important or more frequent than others, you can use
`average="weighted"` to weight each label by its support.


### Classifier Chains

Some classifiers do not natively support multilabel classification (for example,
SVC). One strategy is to train one classifier per label, but this ignores
dependencies between labels.

A better approach is to use *classifier chains*, where each classifier receives
the predictions of previous classifiers as additional inputs.


```python
from sklearn.multioutput import ClassifierChain
from sklearn.svm import SVC
```

We create a classifier chain using cross-validation to generate clean
(out-of-sample) predictions for training.


```python
chain_clf = ClassifierChain(
    SVC(),
    cv=3,
    random_state=42
)

chain_clf.fit(X_train[:2000], y_multilabel[:2000])
```

We can now use the trained classifier chain to make predictions.


```python
chain_clf.predict([some_digit])
```

# Multioutput Classification

The final type of classification task we will cover is **multioutput–multiclass
classification**, often shortened to *multioutput classification*.

This is a generalization of multilabel classification where **each output label
can itself be multiclass**, meaning it can take more than two possible values.


To illustrate this, we will build a system that removes noise from images.

The model will:
- Take a **noisy digit image** as input
- Output a **clean digit image**, represented as pixel intensities

Each output contains one label per pixel, and each pixel can take multiple
values (from 0 to 255). This makes it a **multioutput classification system**.


> **Note**
>
> The distinction between classification and regression can be blurry.
> Predicting pixel intensities looks like regression, but the model still
> predicts multiple discrete outputs per instance.
>
> Multioutput systems are not limited to classification: they can output a mix
> of class labels and numeric values.


```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

We start by creating noisy versions of the MNIST images.

- The inputs will be noisy images
- The targets will be the original, clean images


```python
rng = np.random.default_rng(seed=42)

noise_train = rng.integers(0, 100, (len(X_train), 784))
X_train_mod = X_train + noise_train

noise_test = rng.integers(0, 100, (len(X_test), 784))
X_test_mod = X_test + noise_test

y_train_mod = X_train
y_test_mod = X_test
```

Let’s visualize a noisy test image alongside its clean target.

(Yes, this technically peeks at the test set — in a real project, you should
avoid this!)


```python
plt.subplot(1, 2, 1)
plot_digit(X_test_mod[0])
plt.title("Noisy input")

plt.subplot(1, 2, 2)
plot_digit(y_test_mod[0])
plt.title("Clean target")

plt.show()
```

Now we train a classifier that can handle multioutput predictions.

The KNeighborsClassifier works well here because it can predict multiple output
values per instance.


```python
knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train_mod, y_train_mod)
```

Let’s use the trained model to clean up a noisy digit.


```python
clean_digit = knn_clf.predict([X_test_mod[0]])

plot_digit(clean_digit)
plt.show()
```

The output image is very close to the original clean digit.

## 2. Data Augmentation

Exercise: _Write a function that can shift an MNIST image in any direction (left, right, up, or down) by one pixel. You can use the `shift()` function from the `scipy.ndimage` module. For example, `shift(image, [2, 1], cval=0)` shifts the image two pixels down and one pixel to the right. Then, for each image in the training set, create four shifted copies (one per direction) and add them to the training set. Finally, train your best model on this expanded training set and measure its accuracy on the test set. You should observe that your model performs even better now! This technique of artificially growing the training set is called _data augmentation_ or _training set expansion_._

Let's try augmenting the MNIST dataset by adding slightly shifted versions of each image.

```python
from scipy.ndimage import shift
```

```python
def shift_image(image, dx, dy):
    image = image.reshape((28, 28))
    shifted_image = shift(image, [dy, dx], cval=0, mode="constant")
    return shifted_image.reshape([-1])
```

Let's see if it works:

```python
image = X_train[1000]  # some random digit to demo
shifted_image_down = shift_image(image, 0, 5)
shifted_image_left = shift_image(image, -5, 0)

plt.figure(figsize=(12, 3))
plt.subplot(131)
plt.title("Original")
plt.imshow(image.reshape(28, 28),
           interpolation="nearest", cmap="Greys")
plt.subplot(132)
plt.title("Shifted down")
plt.imshow(shifted_image_down.reshape(28, 28),
           interpolation="nearest", cmap="Greys")
plt.subplot(133)
plt.title("Shifted left")
plt.imshow(shifted_image_left.reshape(28, 28),
           interpolation="nearest", cmap="Greys")
plt.show()
```

Looks good! Now let's create an augmented training set by shifting every image left, right, up and down by one pixel:

```python
X_train_augmented = [image for image in X_train]
y_train_augmented = [label for label in y_train]

for dx, dy in ((-1, 0), (1, 0), (0, 1), (0, -1)):
    for image, label in zip(X_train, y_train):
        X_train_augmented.append(shift_image(image, dx, dy))
        y_train_augmented.append(label)

X_train_augmented = np.array(X_train_augmented)
y_train_augmented = np.array(y_train_augmented)
```

Let's shuffle the augmented training set, or else all shifted images will be grouped together:

```python
rng = np.random.default_rng(seed=42)
shuffle_idx = rng.permutation(len(X_train_augmented))
X_train_augmented = X_train_augmented[shuffle_idx]
y_train_augmented = y_train_augmented[shuffle_idx]
```

Now let's train the model using the best hyperparameters we found in the previous exercise:

```python
knn_clf = KNeighborsClassifier(**grid_search.best_params_)
```

```python
knn_clf.fit(X_train_augmented, y_train_augmented)
```

**Warning**: the following cell may take a few minutes to run:

```python
augmented_accuracy = knn_clf.score(X_test, y_test)
augmented_accuracy
```

By simply augmenting the data, we've got a 0.5% accuracy boost. Perhaps it does not sound so impressive, but it actually means that the error rate dropped significantly:

```python
error_rate_change = (1 - augmented_accuracy) / (1 - tuned_accuracy) - 1
print(f"error_rate_change = {error_rate_change:.0%}")
```

The error rate dropped quite a bit thanks to data augmentation.

```python
def get_email_structure(email):
    if isinstance(email, str):
        return email
    payload = email.get_payload()
    if isinstance(payload, list):
        multipart = ", ".join([get_email_structure(sub_email)
                               for sub_email in payload])
        return f"multipart({multipart})"
    else:
        return email.get_content_type()
```

Okay, let's start writing the preprocessing functions. First, we will need a function to convert HTML to plain text. Arguably the best way to do this would be to use the great [BeautifulSoup](https://www.crummy.com/software/BeautifulSoup/) library, but I would like to avoid adding another dependency to this project, so let's hack a quick & dirty solution using regular expressions (at the risk of [un̨ho͞ly radiańcé destro҉ying all enli̍̈́̂̈́ghtenment](https://stackoverflow.com/a/1732454/38626)). The following function first drops the `<head>` section, then converts all `<a>` tags to the word HYPERLINK, then it gets rid of all HTML tags, leaving only the plain text. For readability, it also replaces multiple newlines with single newlines, and finally it unescapes html entities (such as `&gt;` or `&nbsp;`):

```python
import re
from html import unescape

def html_to_plain_text(html):
    text = re.sub(r'<head.*?>.*?</head>', '', html, flags=re.M | re.S | re.I)
    text = re.sub(r'<a\s.*?>', ' HYPERLINK ', text, flags=re.M | re.S | re.I)
    text = re.sub(r'<.*?>', '', text, flags=re.M | re.S)
    text = re.sub(r'(\s*\n)+', '\n', text, flags=re.M | re.S)
    return unescape(text)
```

Great! Now let's write a function that takes an email as input and returns its content as plain text, whatever its format is:

```python
def email_to_text(email):
    html = None
    for part in email.walk():
        ctype = part.get_content_type()
        if not ctype in ("text/plain", "text/html"):
            continue
        try:
            content = part.get_content()
        except: # in case of encoding issues
            content = str(part.get_payload())
        if ctype == "text/plain":
            return content
        else:
            html = content
    if html:
        return html_to_plain_text(html)
```

We are ready to put all this together into a transformer that we will use to convert emails to word counters. Note that we split sentences into words using Python's `split()` method, which uses whitespaces for word boundaries. This works for many written languages, but not all. For example, Chinese and Japanese scripts generally don't use spaces between words, and Vietnamese often uses spaces even between syllables. It's okay in this exercise, because the dataset is (mostly) in English.

```python
from sklearn.base import BaseEstimator, TransformerMixin

class EmailToWordCounterTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, strip_headers=True, lower_case=True,
                 remove_punctuation=True, replace_urls=True,
                 replace_numbers=True, stemming=True):
        self.strip_headers = strip_headers
        self.lower_case = lower_case
        self.remove_punctuation = remove_punctuation
        self.replace_urls = replace_urls
        self.replace_numbers = replace_numbers
        self.stemming = stemming
    def fit(self, X, y=None):
        return self
    def transform(self, X, y=None):
        X_transformed = []
        for email in X:
            text = email_to_text(email) or ""
            if self.lower_case:
                text = text.lower()
            if self.replace_urls and url_extractor is not None:
                urls = list(set(url_extractor.find_urls(text)))
                urls.sort(key=lambda url: len(url), reverse=True)
                for url in urls:
                    text = text.replace(url, " URL ")
            if self.replace_numbers:
                text = re.sub(r'\d+(?:\.\d*)?(?:[eE][+-]?\d+)?', 'NUMBER', text)
            if self.remove_punctuation:
                text = re.sub(r'\W+', ' ', text, flags=re.M)
            word_counts = Counter(text.split())
            if self.stemming and stemmer is not None:
                stemmed_word_counts = Counter()
                for word, count in word_counts.items():
                    stemmed_word = stemmer.stem(word)
                    stemmed_word_counts[stemmed_word] += count
                word_counts = stemmed_word_counts
            X_transformed.append(word_counts)
        return np.array(X_transformed)
```

Now we have the word counts, and we need to convert them to vectors. For this, we will build another transformer whose `fit()` method will build the vocabulary (an ordered list of the most common words) and whose `transform()` method will use the vocabulary to convert word counts to vectors. The output is a sparse matrix.

```python
from scipy.sparse import csr_matrix

class WordCounterToVectorTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, vocabulary_size=1000):
        self.vocabulary_size = vocabulary_size
    def fit(self, X, y=None):
        total_count = Counter()
        for word_count in X:
            for word, count in word_count.items():
                total_count[word] += min(count, 10)
        most_common = total_count.most_common()[:self.vocabulary_size]
        self.vocabulary_ = {word: index + 1
                            for index, (word, count) in enumerate(most_common)}
        return self
    def transform(self, X, y=None):
        rows = []
        cols = []
        data = []
        for row, word_count in enumerate(X):
            for word, count in word_count.items():
                rows.append(row)
                cols.append(self.vocabulary_.get(word, 0))
                data.append(count)
        return csr_matrix((data, (rows, cols)),
                          shape=(len(X), self.vocabulary_size + 1))
```
